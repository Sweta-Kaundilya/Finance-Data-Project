```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = yf.download('BAC',start=start,end=end)
```

```python
symbols = ['BAC','C','GS','JPM','MS','WFC']
for i in range(6):
    symbols[i] = yf.download(symbols[i],start=start,end=end)
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ('BAC','C','GS','JPM','MS','WFC')
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat(symbols, axis=1, join = 'outer')
```

```python
bank_stocks.head()
```

** Set the column name levels (this is filled out for you):**

```python
bank_stocks.columns.names = ['Stock Info','Bank Ticker']
```

```python
bank_stocks.head()
```

** What is the max Close price for each bank's stock throughout the time period?**

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
percentage = bank_stocks.xs(key='Close', axis=1, level='Stock Info').pct_change()
percentage.head()
```

```python
for i in tickers:
    returns[i + ' Return'] = percentage[i]

returns.head()
```

```python
returns['BAC Return'].idxmax()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
bac2008 = BAC['Close'].loc['2008-01-01':'2009-01-01']
bac2008.rolling(window=30).mean().plot(label = '30 day average', color = 'blue',figsize=(15,5)) # Plot of 30 day MA in 2008
bac2008.plot(label = 'BAC Close', figsize = (15,5), color = 'green')                            # Plot of BAC Close in 2008
plt.legend(loc = 'best')
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
correlations = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
correlations.iplot(kind='heatmap',colorscale = 'rdylbu')
# One can also use : sns.heatmap(correlations, options)
```

```python
sns.heatmap(correlations, annot=True)
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(correlations, annot=True)
```
